Clarifying the concept of *smoothness* for linear regression problems.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-paper')
plt.rc('text', usetex=False)
```

# 0. Generate noisy data

```python
from sklearn import datasets

n_samples = 100

X, y, coef = datasets.make_regression(n_samples=n_samples, n_features=1,
                                      n_informative=1, noise=100,
                                      coef=True, random_state=42)
print('True coef is {}'.format(coef))
```

# 1. Ridge regression

```python
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.model_selection import GridSearchCV

ols = LinearRegression()
ols.fit(X, y)

ridge = GridSearchCV(Ridge(), param_grid={'alpha': np.logspace(0, 2, 50)})
ridge.fit(X, y)

print(ridge.best_estimator_.alpha)
print('w_ols is {}'.format(ols.coef_[0]))
print('w_ridge {}'.format(ridge.best_estimator_.coef_[0]))

xaxis = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)

plt.figure(dpi=100)

# plot data
p1 = plt.scatter(X, y,
                 s=50, linewidth=0.5, edgecolor='k', label='noisy data',
                 color='C0', alpha=1)

# plot prediction
p2, = plt.plot(xaxis, ols.predict(xaxis),
               linestyle='--', color='C1',
               linewidth=3)

p3, = plt.plot(xaxis, ridge.predict(xaxis),
               linestyle='-.', color='C2',
               linewidth=3)


plt.legend([p1, p2, p3],
           ['noisy data',
            r'$\hat w_{OLS}=$'+str(ols.coef_[0])[:6],
            r'$\hat w_{ridge}=$'+str(ridge.best_estimator_.coef_[0])[:6]])
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.title(r'$w^* = {:2.3f}$'.format(np.float(coef)))
#plt.xticks([])
#plt.yticks([])

plt.tight_layout();
```
