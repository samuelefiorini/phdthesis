Popular loss functions.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-paper')
plt.rc('text', usetex=False)
```

### Regression losses

```python
xmin, xmax = -4, 4
xx = np.linspace(xmin, xmax, 100)
lw = 2

plt.figure(dpi=300)

# Square
plt.plot(xx, (xx)**2, lw=lw, color='C2',
         label="Square", ls='-')

# Absolute
plt.plot(xx, np.abs(xx), lw=lw, color='C7',
         label="Absolute", ls='-')

# eps-insensitive
plt.plot(xx, np.where(np.abs(xx) < 1, 0, np.abs(xx) - 1), lw=lw,
         label=r"$\epsilon$-insensitive", ls='-', color='C4')



plt.ylim((-0.1, 6))
legend = plt.legend(loc=1)
legend.get_frame().set_alpha(1)
plt.xlabel(r"$y - f(X)$")
plt.ylabel("$V(f(X), y)$")
plt.savefig('../images/regression_losses.png');
```

### Classification losses

```python
xmin, xmax = -4, 4
xx = np.linspace(xmin, xmax, 100)
lw = 2

plt.figure(dpi=300)

# 0-1
plt.plot([xmin, 0, 0, xmax], [1, 1, 0, 0], lw=lw, color='C1',
         label="Zero-one", ls='-')

# Square
plt.plot(xx, (1-xx)**2, lw=lw, color='C2',
         label="Square", ls='-')

# Logistic
plt.plot(xx, np.log(1 + np.exp(-xx)), lw=lw, color='C3',
         label="Logistic", ls='-')

# Hinge
plt.plot(xx, np.where(xx < 1, 1 - xx, 0), lw=lw,
         label="Hinge", ls='-', color='C4')

# Exponential
plt.plot(xx, np.exp(-xx), lw=lw,
         label="Exponential", ls='-', color='C5')


# plt.plot(xx, np.where(xx < 1, 1 - xx, 0) ** 2, lw=lw,
#          label="Squared hinge loss")
# plt.plot(xx, -np.minimum(xx, 0), lw=lw,
#          label="Perceptron loss")

plt.ylim((-0.1, 6))
legend = plt.legend(loc=1)
legend.get_frame().set_alpha(1)
plt.xlabel(r"$yf(X)$")
plt.ylabel("$V(f(X), y)$")
plt.savefig('../images/classification_losses.png');
```
