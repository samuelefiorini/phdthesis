## Figure 4.5
Square loss contour lines.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-paper')
plt.rc('text', usetex=False)
```

## Ordinary least squares

```python
plt.figure(figsize=(6, 6), dpi=300)

# Contour boundaries
xmax = 2
ymax = 4
xmin = -4
ymin = -2
xshift = 1.5
yshift = - 3.5

delta = 0.01
xaxis = np.arange(2*xmin, 2*xmax, delta)
yaxis = np.arange(2*ymin, 2*ymax, delta)
xx, yy = np.meshgrid(xaxis, yaxis)

# Loss function
loss = (xx + xshift)**2 + (yy + yshift)**2

plt.contour(xx, yy, loss, np.arange(0, 3.5, 0.25), cmap='Blues', zorder=1) # loss
plt.contourf(xx, yy, loss, np.arange(0, 3.5, 0.25),
             cmap='Blues', zorder=1, alpha=0.35,) # loss

plt.scatter(-xshift, -yshift, marker='X', s=90, label=r'$\hat w_{OLS}$')

plt.axhline(0, -1, 1, c='gray', zorder=0)
plt.axvline(0, -1, 1, c='gray', zorder=0)
plt.xlim([xmin, xmax])
plt.ylim([ymin, ymax])
plt.xlabel(r'$w_1$')
plt.ylabel(r'$w_2$')
plt.xticks([])
plt.yticks([])
plt.legend()

plt.savefig('../images/w_ols.png');
```

## Ridge Regression

```python
plt.figure(figsize=(6, 6), dpi=300)

# Contour boundaries
xmax = 2
ymax = 4
xmin = -4
ymin = -2
xshift = 1.5
yshift = - 3.5

delta = 0.01
xaxis = np.arange(2*xmin, 2*xmax, delta)
yaxis = np.arange(2*ymin, 2*ymax, delta)
xx, yy = np.meshgrid(xaxis, yaxis)

loss = (xx + xshift)**2 + (yy + yshift)**2 # Loss function
penalty = xx**2 + yy**2 # L2-penalty

plt.contour(xx, yy, loss, np.arange(0, 3.5, 0.25), cmap='Blues', zorder=1) # loss
plt.contourf(xx, yy, loss, np.arange(0, 3.5, 0.25),
             cmap='Blues', zorder=1, alpha=0.35) # loss

plt.contour(xx, yy, penalty, [4], cmap='Reds_r', zorder=2) # L2-penalty

p1 = plt.scatter(-xshift, -yshift, marker='X', s=90, zorder=2)
p2 = plt.scatter(-0.8, 1.85, marker='o', s=90, zorder=2, c='firebrick')

plt.axhline(0, -1, 1, c='gray', zorder=0)
plt.axvline(0, -1, 1, c='gray', zorder=0)

p3, = plt.plot([0, -2], [0, 0], c='firebrick', linestyle='dashed', linewidth=3)

plt.xlim([xmin, xmax])
plt.ylim([ymin, ymax])
plt.xlabel(r'$w_1$')
plt.ylabel(r'$w_2$')
plt.xticks([])
plt.yticks([])
plt.legend([p1, p2, p3], [r'$\hat w_{OLS}$', r'$\hat w_{\ell_2}$', r'$\propto \lambda^{-1}$'])

plt.savefig('../images/w_ridge.png');
```

## Lasso

```python
plt.figure(figsize=(6, 6), dpi=300)

# Contour boundaries
xmax = 2
ymax = 4
xmin = -4
ymin = -2
xshift = 1.5
yshift = - 3.5

delta = 0.01
xaxis = np.arange(2*xmin, 2*xmax, delta)
yaxis = np.arange(2*ymin, 2*ymax, delta)
xx, yy = np.meshgrid(xaxis, yaxis)

loss = (xx + xshift)**2 + (yy + yshift)**2 # Loss function
penalty = np.abs(xx) + np.abs(yy) # L1-penalty

plt.contour(xx, yy, loss, np.arange(0, 4.75, 0.25), cmap='Blues', zorder=1) # loss
plt.contourf(xx, yy, loss, np.arange(0, 4.75, 0.25),
             cmap='Blues', zorder=1, alpha=0.35) # loss

#plt.contour(xx, yy, penalty, [2], cmap='Reds_r', zorder=0) # L2-penalty
plt.contour(xx, yy, penalty, [2], cmap='Greens_r', zorder=2) # L1-penalty

p1 = plt.scatter(-xshift, -yshift, marker='X', s=90, zorder=1)
p2 = plt.scatter(0, 2, marker='p', s=90, zorder=2, c='darkgreen')

plt.axhline(0, -1, 1, c='gray', zorder=0)
plt.axvline(0, -1, 1, c='gray', zorder=0)

p3, = plt.plot([0, -2], [0, 0], c='darkgreen', linestyle='dashed', linewidth=3)

plt.xlim([xmin, xmax])
plt.ylim([ymin, ymax])
plt.xlabel(r'$w_1$')
plt.ylabel(r'$w_2$')
plt.xticks([])
plt.yticks([])
plt.legend([p1, p2, p3], [r'$\hat w_{OLS}$', r'$\hat w_{\ell_1}$', r'$\propto \lambda^{-1}$'])

plt.savefig('../images/w_lasso.png');
```

## Elastic-Net

```python
plt.figure(figsize=(6, 6), dpi=300)

# Contour boundaries
xmax = 2
ymax = 4
xmin = -4
ymin = -2
xshift = 1.5
yshift = - 3.5

delta = 0.01
xaxis = np.arange(2*xmin, 2*xmax, delta)
yaxis = np.arange(2*ymin, 2*ymax, delta)
xx, yy = np.meshgrid(xaxis, yaxis)

loss = (xx + xshift)**2 + (yy + yshift)**2 # Loss function
penalty_l1 = np.abs(xx) + np.abs(yy) # L1-penalty
penalty_l2 = xx**2 + yy**2 # L2-penalty
penalty_enet = penalty_l1 + penalty_l2

plt.contour(xx, yy, loss, np.arange(0, 4, 0.25), cmap='Blues', zorder=1) # loss
plt.contourf(xx, yy, loss, np.arange(0, 4, 0.25),
             cmap='Blues', zorder=1, alpha=0.35) # loss

plt.contour(xx, yy, penalty_enet, [6], cmap='Purples_r', zorder=2) # enet-penalty

p1 = plt.scatter(-xshift, -yshift, marker='X', s=90, zorder=1)
p2 = plt.scatter(-0.6, 1.8, marker='h', s=90, zorder=2, c='purple')

plt.axhline(0, -1, 1, c='gray', zorder=0)
plt.axvline(0, -1, 1, c='gray', zorder=0)

plt.xlim([xmin, xmax])
plt.ylim([ymin, ymax])
plt.xlabel(r'$w_1$')
plt.ylabel(r'$w_2$')
plt.xticks([])
plt.yticks([])
plt.legend([p1, p2], [r'$\hat w_{OLS}$', r'$\hat w_{\ell_1\ell_2}$'])

plt.savefig('../images/w_enet.png');
```

```python
plt.figure(figsize=(6, 6), dpi=300)

# Contour boundaries
xmax = 2
ymax = 4
xmin = -4
ymin = -2
xshift = 1.5
yshift = - 3.5

delta = 0.01
xaxis = np.arange(2*xmin, 2*xmax, delta)
yaxis = np.arange(2*ymin, 2*ymax, delta)
xx, yy = np.meshgrid(xaxis, yaxis)

loss = (xx + xshift)**2 + (yy + yshift)**2 # Loss function
penalty_l1 = np.abs(xx) + np.abs(yy) # L1-penalty
penalty_l2 = xx**2 + yy**2 # L2-penalty
penalty_enet = penalty_l1 + penalty_l2

plt.contour(xx, yy, loss, np.arange(0, 4.75, 0.25), cmap='Blues', zorder=1) # loss
plt.contourf(xx, yy, loss, np.arange(0, 4.75, 0.25),
             cmap='Blues', zorder=1, alpha=0.35) # loss

plt.contour(xx, yy, penalty_l2, [4], cmap='Reds_r', zorder=2) # L2-penalty
plt.contour(xx, yy, penalty_l1, [2], cmap='Greens_r', zorder=2) # L1-penalty
plt.contour(xx, yy, penalty_enet, [6], cmap='Purples_r', zorder=2) # enet-penalty

p1 = plt.scatter(-xshift, -yshift, marker='X', s=90, zorder=2)
p2 = plt.scatter(-0.6, 1.8, marker='h', s=90, zorder=2, c='purple') # w_enet
p3 = plt.scatter(0, 2, marker='p', s=90, zorder=2, c='darkgreen') # w_lasso
p4 = plt.scatter(-0.8, 1.85, marker='o', s=90, c='firebrick', zorder=2) # w_ridge


plt.axhline(0, -1, 1, c='gray', zorder=0)
plt.axvline(0, -1, 1, c='gray', zorder=0)

plt.xlim([xmin, xmax])
plt.ylim([ymin, ymax])
plt.xlabel(r'$w_1$')
plt.ylabel(r'$w_2$')
plt.xticks([])
plt.yticks([])
plt.legend([p1, p4, p3, p2], [r'$\hat w_{OLS}$', r'$\hat w_{\ell_2}$',
                              r'$\hat w_{\ell_1}$',
                              r'$\hat w_{\ell_1\ell_2}$'])

plt.savefig('../images/w_all.png');
```
